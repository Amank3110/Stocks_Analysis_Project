```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
directory = r'C:\Users\Aman\Documents\Stock-Analysis-Project\S&P_resources\individual_stocks_5yr'
company = ['GOOG_data','MSFT','AMZN','AAPL']

files = [files for files in os.listdir(directory) if files.endswith('.csv') if any(name in files for name in company)]
df=pd.DataFrame()
for file in files:
    dff = pd.read_csv(os.path.join(directory,file),error_bad_lines=False)
    df = pd.concat([df,dff],ignore_index=True)
```

```python
df['date']=pd.to_datetime(df['date'])
```

# Stock Price Change over time!

```python
company_list = df['Name'].unique()
```

```python
plt.figure(figsize=(15,10))
for index,company in enumerate(company_list,1):
    plt.subplot(2,2,index)
    filter1 = df['Name']==company
    new_df=df[filter1]
    plt.plot(new_df['date'],new_df['close'])
    plt.title(company)
    
```

# Analyse Moving average of multiple stocks

```python
new_df[new_df['date']=='2018-02-02']
```

```python
days = [10,20,50]
new_df = df.copy()
for i in days:
    new_df[f'close_{i}'] = new_df['close'].rolling(i).mean()
plt.figure(figsize=(20,15))
for index,company in enumerate(company_list,1):
    plt.subplot(2,2,index)
    filter1 = new_df['Name']==company
    new_df1=new_df[filter1]
    new_df1[['close_10','close_20','close_50']].plot(ax=plt.gca())
    plt.title(company)
```
